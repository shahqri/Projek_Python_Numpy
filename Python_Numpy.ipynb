```python
import numpy as np

# Buat array 2D
arr_2d = np.array([[1,2,3], [4,5,6]])

# Operasi matematik pantas!
print("Bentuk (Shape):", arr_2d.shape) # Output : (2,3)
print("Jumlah Paksi 0 :", np.sum(arr_2d, axis = 0))
print("Jumlah Paksi 1 :", np.sum(arr_2d, axis = 1))

# Buat senarai nombor automatik
nombor = np.arange(24).reshape(2, 3, 4)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100) # Guna Numpy
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots() # Buat kanvas
ax.plot(x, y1, color = 'blue', label = 'Sin')
ax.plot(x, y2, color = 'red', label = 'Cos')

ax.set_title("Perbandingan Sin & Cos")
ax.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
nama_buah = ['Epal', 'Oren', 'Lemon']
jualan = [10, 15, 5]

# Carta bar
plt.bar(nama_buah, jualan)


# Tajuk dan label paksi
plt.title("Jualan Buah Hari Ini")
plt.xlabel("Nama Buah")
plt.ylabel("Bilangan Jualan")

# Paparan graf
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
nama_buah = ['Epal', 'Oren', 'Lemon']
jualan = [10, 15, 5]

# Carta bar dengan warna
warna = ['crimson', 'darkorange', 'gold']
plt.bar(nama_buah, jualan, color=warna)

# Tajuk dan label paksi
plt.title("Jualan Buah Hari Ini")
plt.xlabel("Nama Buah")
plt.ylabel("Bilangan Jualan")

# Hadkan paksi-Y ke 16 supaya angka 15 ada ruang di atas & tak terpotong
plt.ylim(0, 16)

# Memaparkan sukatan paksi-Y satu per satu dari 0 hingga 15
plt.yticks(range(0, 16))

# Paparan graf
plt.show()
```
